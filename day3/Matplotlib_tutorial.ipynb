# Line Plot

```python
#line plot 
plt.style.use('fivethirtyeight')
x = [1,2,3,4,5]
y= [6,8,9,11,12]

x2 = [1,2,3,4,5]
y2= [1,4,9,16,25]
plt.figure(figsize=(8,6))
plt.title("Sample Graph",fontdict = {'fontsize' : 15})
plt.plot(x,y, color='red',label="Relation between x and y", linewidth=3 , linestyle='--',marker='.',markersize=12, markeredgecolor='blue')
plt.plot(x2 , y2 ,'b-^',label="Square relation")
plt.xlabel("Horizontal Axis ")
plt.ylabel("Vertical Axis")
plt.legend()
plt.savefig("mygraph.png",dpi=100)

plt.show()



```

# Bar chart

```python
# plt.style.use('seaborn')
label = ["India ", "USA" , "China"]
value = [4,2,1]
bar = plt.bar(label , value,color="y")
patterns = ["O" , "*" , "/"]
plt.xlabel("Country Name")
plt.title("Country Comparison")
plt.ylabel("Ranking")
for b in bar:
    b.set_hatch(patterns.pop())
#set_hatch
# bar.set_hatch('O')
plt.savefig("barchart.png")
plt.show()
```

# Scatter Plot

```python
x= [1,2,3,4,8,9,5,6]
y = [2,3,7,8,9,4,2,0]

plt.scatter(x,y, c='r', marker='^',label="Triangle")
plt.scatter([0,1,2],[4,5,6],c='green',label="Circles")
plt.legend()
plt.savefig('scatter.png')
plt.show()
```
